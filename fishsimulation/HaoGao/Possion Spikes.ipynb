```python
def gen_spikes(f):
    """
    
    :param f: firing rate
    :return: 
    """

    s = torch.ones((10, 10000)).bernoulli_(f / 1000)
    w = torch.rand((4, 10))
    w[:2, 8:] = 0
    w[2:, :8] = 0

    return (w @ s).T.unsqueeze(dim=2)
```

```python
sp = gen_spikes(20)
```
